```python
# Проверим, влияет ли уровень экологической обстановки в районе на цену квартиры. Постройте коробчатую диаграмму цен на квартиры (price_doc) в зависимости от уровня экологической обстановки в районе (ecology). Какой уровень ценится на рынке меньше всего?

# Загрузка данных
sber_data = pd.read_csv('https://raw.githubusercontent.com/PurunSA/my_projects/refs/heads/main/Sber/sber_data.csv')

# Проверим уникальные значения в колонке ecology
print(sber_data['ecology'].value_counts())

# Построим boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='ecology', y='price_doc', data=sber_data)
plt.title('Распределение цен на квартиры в зависимости от уровня экологии')
plt.xlabel('Уровень экологической обстановки')
plt.ylabel('Цена квартиры')
plt.grid(True)
plt.show()
```

```python
#display(sber_data.isnull().tail()) -- метод isnull() возвращает новый DataFrame, в ячейках которого стоят булевы значения True и False. True ставится на месте, где ранее находилось значение NaN

# вычислим средний по столбцам результат метода isnull(). Получим долю пропусков в каждом столбце. Умножаем на 100 %, находим столбцы, где доля пропусков больше 0, сортируем по убыванию
cols_null_percent = sber_data.isnull().mean() * 100
cols_with_null = cols_null_percent[cols_null_percent>0].sort_values(ascending=False)
display(cols_with_null) 
```
